# Project: Chest X-Ray Classification

#### Project Goal: 
Build a deep learning model to classify chest X-rays as either COVID-19, viral pneumonia, or normal.

#### Data: 
Chest X-ray images obtained from kaggle at https://www.kaggle.com/pranavraikokte/covid19-image-dataset. Separate train and test data are provided, so no need to split the data.

#### Analysis: 
Keras deep learning model will be implemented and model performance over epochs will be visualized in order to determine if model is satisfactory.

#### Evaluation: 
The project will conclude with the evaluation of the deep learning model on the test data set. The output of the predictions will be checked with a confusion matrix and metrics such as accuracy, precision, recall, and F1 scores.

```python
# import python modules

import numpy
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import layers

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BATCH = 4

# load/batch training data, preprocess with pixel normalization, and perform data augmentation
training_data_generator = ImageDataGenerator(rescale=1.0/255, zoom_range=0.1, rotation_range=10, width_shift_range=0.05, height_shift_range=0.05)
training_iterator = training_data_generator.flow_from_directory('Covid19-dataset/train',class_mode='categorical',color_mode='grayscale',batch_size=BATCH)

# load/batch test data, preprocess with pixel normalization
validation_data_generator = ImageDataGenerator(rescale=1.0/255)
validation_iterator = validation_data_generator.flow_from_directory('Covid19-dataset/test',class_mode='categorical',color_mode='grayscale',batch_size=BATCH, shuffle=False)
```

```python
# build model architecture with convolutional and max pooling layers 
model = Sequential()
model.add(layers.InputLayer(input_shape=training_iterator.image_shape))
model.add(layers.Conv2D(8, 4, activation="relu")) 
model.add(layers.MaxPooling2D(pool_size=(4,4), strides=(4,4)))
model.add(layers.Conv2D(32, 3, activation="relu"))
model.add(layers.MaxPooling2D(pool_size=(2,2), strides=(2,2)))
model.add(layers.Dropout(0.1))
model.add(layers.Flatten())
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(3, activation='softmax'))
model.summary()

# compile the model with an Adam optimizer and categorical cross entropy loss
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['accuracy', tf.keras.metrics.AUC()])
```

```python
# use early stopping to monitor validation loss
es = EarlyStopping(monitor='val_loss', mode='min', patience=12, verbose=0)
# use this callback to save model with highest validation accuracy
mc = ModelCheckpoint('best_model.h5', monitor='val_accuracy', mode='max', verbose=0, save_best_only=True)

# train and validate model
history = model.fit(training_iterator, epochs=100, validation_data=validation_iterator, callbacks=[es, mc], verbose=0)
```

```python
# plotting training and validation data accuracy over epochs
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(history.history['accuracy'])
ax1.plot(history.history['val_accuracy'])
ax1.set_title('Model Accuracy')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend(['train', 'validation'], loc='lower right')
 
# plotting training and validation data auc over epochs
ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(history.history['auc'])
ax2.plot(history.history['val_auc'])
ax2.set_title('Model AUC')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('AUC')
ax2.legend(['train', 'validation'], loc='lower right')

# plotting training and validation data loss over epochs
ax2 = fig.add_subplot(2, 2, 3)
ax2.plot(history.history['loss'])
ax2.plot(history.history['val_loss'])
ax2.set_title('Model Cross-Entropy Loss')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend(['train', 'validation'], loc='lower right')
 
# keep plots from overlapping
fig.tight_layout()
```

```python
# load saved model with highest validation accuracy and compile
model.load_weights('best_model.h5')
model.compile(optimizer=opt, loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['accuracy', tf.keras.metrics.AUC()])

# determine and print classification report
predictions = model.predict(validation_iterator)
predicted_classes = numpy.argmax(predictions, axis=1)
true_classes = validation_iterator.classes
class_labels = list(validation_iterator.class_indices.keys())
report = classification_report(true_classes, predicted_classes, target_names=class_labels)
print(report)   
 
# plot heatmap of confusion matrix
cm=confusion_matrix(true_classes,predicted_classes)
plt.figure(figsize = (10, 8))
ax = sns.heatmap(cm, annot=True, cmap='Blues')
ax.set_yticklabels(labels = class_labels, va='center')
ax.set_xticklabels(labels = class_labels, va='center')
plt.title('Confusion Matrix');
```

The numbers along the diagonal of the heatmap show the correct classifications. All other numbers show mis-classifications. It can be seen from the classification report that the model achieved an accuracy of 0.97, with recalls for the COVID-19, normal, and viral pneumonia scans as 1.0, 1.0, and 0.9, respectively. The model appears to be satisfactory with these results greater than or equal to 0.90. The recall metric is of more concern than precision for the disease cases. This is due to the fact that false negatives for cases with disease should be minimized as it may be detrimental to conclude that there is no disease when in fact disease is present.
